```python
def fill_salary(row):
    start_salary = np.nan
    end_salary = np.nan
    
    if pd.isnull(row['Начальная зарплата']) or pd.isnull(row['Конечная зарплата']):
        if row['Требуемый опыт'] in ['от 3 лет до 6 лет', 'Более 6 лет']:
            start_salary = np.around(np.random.randint(80000, 100000) / 5000) * 5000
            end_salary = np.around(np.random.randint(100000, 250000) / 5000) * 5000
        elif row['Требуемый опыт'] in ['Нет опыта', 'От 3 до 6 лет']:
            start_salary = np.around(np.random.randint(20000, 50000) / 5000) * 5000
            end_salary = np.around(np.random.randint(80000, 100000) / 5000) * 5000
    
    return start_salary, end_salary

# Применяем функцию к датафрейму для заполнения начальной и конечной зарплаты
df[['Начальная зарплата', 'Конечная зарплата']] = df.apply(lambda row: fill_salary(row), axis=1, result_type='expand')

print(df)
```

```python
# Функция для заполнения столбца "График работы" с пустыми значениями
def fill_work_schedule(row):
    if pd.isnull(row['График работы']):
        return np.random.choice(time_schedule)
    else:
        return row['График работы']

# Применяем функцию к датафрейму для заполнения столбца "График работы"
df['График работы'] = df.apply(fill_work_schedule, axis=1)

df
```

```python
import re
import pandas as pd
from nltk.tokenize import word_tokenize

df = pd.read_excel('clear_dataframe.xlsx')

# Применение регулярного выражения и токенизации к каждой строке
def process_text(text):
    processed_text = re.sub('<[^>]*>|[а-яё.,!?:\n</>()\d&\[\]\-]', '', text, flags=re.I)  # Добавлены символы '\d&\[\]\-'
    tokens = word_tokenize(processed_text)
    return tokens

# Применение функции к столбцу "Ключевые навыки"
df['Ключевые навыки'] = df['Ключевые навыки'].apply(process_text)

# Сохранение данных
df.to_excel('clear_tokenized_dataframe.xlsx', index=False)
```
